```python
from matplotlib import pyplot as plt


def plot_something() -> plt.Figure:
    # Step 2: Create data
    x = [1, 2, 3, 4, 5]
    y = [2, 4, 6, 8, 10]

    # Step 3: Create a figure and axes
    fig, ax = plt.subplots()

    # Step 4: Plot the data on the axes
    ax.plot(x, y, label="Linear Function")

    # Step 5: Customize the plot
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")
    ax.set_title("Basic Matplotlib Plot")
    ax.legend()

    return fig


figure = plot_something()
figure2 = plot_something()

print(f"Type of the Figure is: {type(figure)}")
print(
    f"Is matplotlib.figure.Figure is the same type as plt.Figure? {isinstance(figure, plt.Figure)}"
)
print(f"id of figure: {id(figure)}")
print(f"id of figure2: {id(figure2)}")
print(f"{figure == figure2}")
print(f"{figure is figure2}")
```

```python
def remove_ticks(axes: plt.Axes) -> plt.Axes:
    axes.tick_params(axis="both", which="both", length=0)
    axes.set_xticks([])
    axes.set_yticks([])
    return axes
```

```python
from contextlib import contextmanager
import matplotlib.pyplot as plt


def apply_style(style_dict):
    """Apply style to matplotlib."""
    plt.rcParams.update(style_dict)


@contextmanager
def styled_figure(styles):
    """Context manager for applying different styles to matplotlib figures."""
    # Save current style
    original_style = plt.rcParams.copy()
    try:
        # Apply desired style
        apply_style(styles)
        # Yield control back to the caller
        yield
    finally:
        # Restore original style
        apply_style(original_style)


axis_styles = [
    {"lines.linestyle": "--", "lines.linewidth": 2, "axes.labelsize": 12},
    {"lines.linestyle": ":", "lines.linewidth": 1.5, "axes.labelsize": 10},
    {"lines.linestyle": "-.", "lines.linewidth": 1, "axes.labelsize": 8},
    {"lines.linestyle": "-", "lines.linewidth": 1.2, "axes.labelsize": 9},
]

# Example of usage
with styled_figure(axis_styles[0]):
    fig, axes = plt.subplots(2, 2)
    for ax, style in zip(axes.flatten(), axis_styles):
        with styled_figure(style):
            ax.plot([1, 2, 3], [1, 2, 3])
    plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams["text.usetex"] = False


def plot(family):
    plt.rcParams["mathtext.fontset"] = family
    fig = plt.figure(figsize=(3, 1.75), dpi=100)
    ax = plt.subplot(
        1, 1, 1, frameon=False, xlim=(-1, 1), ylim=(-0.5, 1.5), xticks=[], yticks=[]
    )
    ax.text(
        0,
        0.0,
        r"$\frac{\pi}{4} = \sum_{k=0}^\infty\frac{(-1)^k}{2k+1}$",
        size=32,
        ha="center",
        va="bottom",
    )
    ax.text(
        0,
        -0.1,
        'mathtext.fontset = "%s"' % family,
        size=14,
        ha="center",
        va="top",
        family="Roboto Condensed",
        color="0.5",
    )
    print(f"after plotter {id(fig)=}")
    print(f"after plotter {id(ax)=}")
    # plt.tight_layout()
    plt.show()


plot("cm")
plot("stix")
plot("stixsans")
plot("dejavusans")
plot("dejavuserif")
plot("custom")
```

```python
def my_plotter(ax: plt.Axes, data1: list, data2: list, param_dict={}) -> plt.Axes:
    """
    A helper function to make a graph.
    """
    ax.plot(data1, data2, **param_dict)
    ax.scatter(data1, data2, **param_dict)
    ax.fill(data1, data2, **param_dict)
    return ax


x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
fig, ax = plt.subplots(1, 1)
ax = my_plotter(ax, x, y)
print(f"after plotter {id(fig)=}")
print(f"after plotter {id(ax)=}")
plt.show()
```
