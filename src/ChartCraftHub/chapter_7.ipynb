```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [4, 5, 6])
print(id(ax))
inspect(ax)
ax.scatter([1, 2, 3], [4, 5, 6])
ax.fill([1, 2, 3], [4, 5, 6])
print(id(ax))
inspect(ax)
plt.show()
```
