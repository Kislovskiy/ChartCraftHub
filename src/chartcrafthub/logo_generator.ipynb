# Logo and rotations

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
X = np.array([3, 5, 1])
Y = np.array([1, 4, 4])
M = np.array([X, Y])

fig, ax = plt.subplots()
ax.scatter(X, Y)
ax.set_xlim(0, 6)
ax.set_ylim(0, 5)
ax.set_aspect("equal")
ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(X, Y, "bo")
ax.plot(X, Y, "bo--")
ax.grid()
ax.set_aspect("equal")
ax.set_xlim(0, 6)
ax.set_ylim(0, 5)
# ax.set_axis_off()
plt.show()
```

```python
from matplotlib.patches import Polygon

fig, ax = plt.subplots()

patch = Polygon(M.T, alpha=0.5)
ax.add_patch(patch)
ax.set_xlim(0, 6)
ax.set_ylim(0, 5)
plt.show()
```

```python
from matplotlib.patches import Polygon
from matplotlib.transforms import Affine2D
from matplotlib.patches import PathPatch

fig, ax = plt.subplots()

for i in range(10):
    transforms = Affine2D().rotate(i * 2 * np.pi / 10)
    transforms.get_matrix()
    transforms.transform_path(patch.get_path())
    ax.add_patch(
        PathPatch(
            transforms.transform_path(patch.get_path()),
            alpha=0.5,
            fill=True,
            color="gray",
        )
    )
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_aspect("equal")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.transforms import Affine2D
from matplotlib.patches import PathPatch


def apply_transforms(patch, ax, cmap, i):
    transforms = Affine2D().translate(-3, -2.93).rotate(i * np.pi / 20)
    transformed_path = transforms.transform_path(patch.get_path())
    ax.add_patch(
        PathPatch(transformed_path, alpha=1, fill=False, color=cmap(i), zorder=-i)
    )


def setup_axes(ax):
    ax.set_aspect("equal")
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_yticks([])
    ax.set_xticks([])
    ax.spines[["top", "right", "bottom", "left"]].set_visible(False)


def plot_chart(ax):
    ax.text(
        0,
        0.5,
        "Chart\nCraft\nHub",
        fontsize=40,
        ha="left",
        va="center",
        font="monospace",
        fontweight="bold",
        color=cmap(2),
    )


def setup_polar_axes(ax, cmap):
    ax_pos = (0.22, 0.4, 0.2, 0.2)
    lw_bars = 1
    ax = fig.add_axes(ax_pos, projection="polar")

    N = 7
    arc = 2.0 * np.pi
    theta = np.arange(0.0, arc, arc / N)
    radii = np.array([2, 6, 8, 7, 4, 5, 8])
    width = np.pi / 4 * np.array([0.4, 0.4, 0.6, 0.8, 0.2, 0.5, 0.3])
    bars = ax.bar(
        theta,
        radii / 6,
        width=width,
        bottom=0.0,
        align="edge",
        edgecolor="lightgray",
        lw=lw_bars,
    )
    for r, bar in zip(radii, bars):
        color = *cm.jet(r / 10.0)[:3], 0.1
        bar.set_facecolor(color)

    ax.tick_params(labelbottom=False, labeltop=False, labelleft=False, labelright=False)
    ax.set_yticks([])
    ax.set_xticks([])


# Main
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6, 4))

N = 8
cmap = plt.get_cmap("bone", N)
for i in range(N):
    apply_transforms(patch, ax[0], cmap, i)

setup_axes(ax[0])
aspect = 2.3
ax[0].set_xlim(-aspect, aspect)
ax[0].set_ylim(-aspect, aspect)
ax[0].set_xlabel(
    "Made with Human Intelligence and Matplotlib",
    loc="center",
    fontsize=6,
    font="monospace",
)
plot_chart(ax[1])
setup_axes(ax[1])
setup_polar_axes(ax[1], cmap)

fig.subplots_adjust(wspace=0, hspace=0)
fig.savefig("results/logo.svg", transparent=True, bbox_inches="tight", dpi=300)
plt.show()
```
